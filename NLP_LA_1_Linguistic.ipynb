Task 1: Tokenization

```python
from nltk.tokenize import sent_tokenize, word_tokenize

sentences = sent_tokenize(review)
print(sentences)
```

```python
words = word_tokenize(review)
print(words)
```

Task 2: Basic Text Analysis

```python
num_characters = len(review)

num_sentences = len(sentences)

num_words = len(words)

avg_sentence_length = num_words / num_sentences

avg_word_length = sum(len(word) for word in words) / num_words

print("Number of characters:", num_characters)
print("Number of sentences:", num_sentences)
print("Number of tokens (words):", num_words)
print("Average sentence length:", avg_sentence_length)
print("Average word length:", avg_word_length)
```

Task 3: Text Normalization

```python
lower_text = review.lower()
print(lower_text)
```

Task 4: POS Tagging

```python
import nltk
from nltk import sent_tokenize, word_tokenize, pos_tag
from nltk.probability import FreqDist

sentences = sent_tokenize(lower_text)

pos_tagged_sentences = [
    pos_tag(word_tokenize(sentence))
    for sentence in sentences
]
print(pos_tagged_sentences[0])
```

Why is POS tagging sensitive to preprocessing choices such as stopword removal?

Removing stopwords or punctuation can disrupt grammatical structure, causing the POS tagger to misclassify words. For example, determiners and prepositions help identify noun phrases, and verbs rely on surrounding words for tense and role disambiguation. Aggressive preprocessing reduces syntactic cues, leading to tagging errors and unreliable downstream tasks like chunking or parsing.

```python
all_tags = [
    tag for sent in pos_tagged_sentences
    for (_, tag) in sent
]

tag_freq = FreqDist(all_tags)
tag_freq.most_common(10)
```

```python
from nltk.chunk import RegexpParser

grammar = r"""
  NP: {<DT>?<JJ>*<NN.*>+}
  VP: {<VB.*><NP|PP>*}
"""
```

```python
chunk_parser = RegexpParser(grammar)

chunked_sentences = [
    chunk_parser.parse(sent)
    for sent in pos_tagged_sentences
]
```
